## 1. Introdução

<p>
A OMC (Organização Mundial do Comércio) é uma instituição internacional que atua na fiscalização e regulamentação do comércio mundial. Com sede em Genebra (Suíça) foi fundada em 1995, durante a Conferência de Marrakech. Fazem parte da OMC 164 países membros (desde julho de 2016). Além disso, a OMC cobre mais de 90% do comércio internacional.
</p>

<p>
Em 11 de Dezembro de 2001, a China tornou-se o 143º membro da OMC. Na época especialistas diziam que a adesão da China à OMC poderia vir a ser tão importante para o século XXI como a revolução comunista foi para o século XX. Anthony Saich, professor convidado na Escola de Políticas Públicas e Gestão da Universidade de Tsinghua, na China, afirmava na época: "Esta adesão vai permitir à China a plena inserção na economia mundial e a aceitação dos princípios de mercado, trazendo consigo modificações extremamente positivas, não só na China, como nas economias que com ela se relacionam".
</p>

<p>
O seguinte projeto tem como objetivo analisar dados e incorporar pesquisas históricas a resultados de relatórios econômicos para verificar se estas previsões abordadas foram coerentes. Além disso, almejo compreender como a China e os países ao seu redor se comportaram diante desta inserção.
</p>

Além disso, a correlação de todas as possibilidades de comparações foram plotadas, o que será uma grande ferramental para a conclusão dos dados. As variáveis serão analisadas separadamente com o IDH.

```python
def reta(dados_x, dados_y, titulo, axis, eixo_x, eixo_y, cor1, cor2, corr):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
 
    plt.scatter(dados_x, dados_y, c=cor1, alpha=0.8)
    plt.axis(axis)
    plt.title(titulo)
    plt.xlabel(eixo_x)
    plt.ylabel(eixo_y)
 
    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=cor2)
    plt.text(72, 0.95, "Correlação: " + str(corr)[:5], color = 'Black')
    plt.grid()
    plt.show()
```
